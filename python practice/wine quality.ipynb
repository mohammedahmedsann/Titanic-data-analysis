```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('winequality-red.csv')
```

```python
title_font = {'family': 'arial', 'color':'darkblue', 'weight' : 'old', 'size' : 15}
axis_font = {'family': 'arial', 'color':'darkblue', 'weight' : 'old', 'size' : 15}
```

```python
for i in df.columns[:-1]:
    plt.figure(figsize = (10, 15))
    sns.barplot(x = 'quality', y= i, data = df)
    
    plt.title('quality-' + str(i), fontdict=title_font)
    plt.xlabel('quality', fontdict = axis_font)
    plt.ylabel(i, fontdict = axis_font)
    
    plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define font dictionaries
title_font = {'family': 'Arial', 'color': 'darkblue', 'weight': 'bold', 'size': 15}
axis_font = {'family': 'Arial', 'color': 'darkblue', 'weight': 'bold', 'size': 15}

# Iterate through columns of the DataFrame (excluding the last column)
for i in df.columns[:-1]:
    plt.figure(figsize=(10, 6))  # Adjust figure size as needed
    sns.barplot(x='quality', y=i, data=df, ci=None)  # `ci=None` disables confidence intervals
    
    # Add title and axis labels
    plt.title('quality - ' + str(i), fontdict=title_font)
    plt.xlabel('quality', fontdict=axis_font)
    plt.ylabel(i, fontdict=axis_font)
    
    # Adjust layout
    plt.tight_layout()
    plt.show()
```
