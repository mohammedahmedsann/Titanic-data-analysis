```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('iris.csv')
```

```python
df.head()
```

```python
df.info()
```

### Knowing the dataset

```python
df.Species.value_counts()
```

```python
df.drop(['Id'], axis = 1, inplace = True)
```

```python
df['Species'] = df['Species'].str.replace('Iris-', '')
```

```python
df.head()
```

```python
plt.figure(figsize = (8,5))
sns.pairplot(df, hue = 'Species', height = 2.75)
```

```python
df['Species_numerical'] = df['Species']
```

```python
df
```

```python
df['Species_numerical'] = df['Species_numerical'].str.replace('setosa', '1')
df['Species_numerical'] = df['Species_numerical'].str.replace('versicolor', '2')
df['Species_numerical'] = df['Species_numerical'].str.replace('virginica', '3')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Species_numerical'] = pd.to_numeric(df['Species_numerical'], downcast = 'integer')
```

```python
df.info()
```

```python
plt.figure(figsize =(5,5))
sns.heatmap(df.corr(numeric_only = True), cmap = 'Spectral', annot = True)
```
